```python
def q2a(r):
    x = sy.Symbol("x")
    return sy.integrate(x**r*sy.exp(-x), (x, 0, oo))
print("EX = ",q2a(1))
print("EX^2 = ",q2a(2))
print("Variance =  ",q2a(2)-q2a(1)**2)
print("EX^r = ",q2a(5))
           
```
