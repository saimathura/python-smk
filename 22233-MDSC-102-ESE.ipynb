## 6. Write a program to test the hypothesis 𝐻0: 𝜇 = 1.75 vs 𝐻1: 𝜇 ≠ 1.75 based on the following observations:1.83, 1.83, 1.73, 1.82, 1.83, 1.73, 1.99, 1.85, 1.68, 1.87

```python
x= [ 1.83, 1.83, 1.73, 1.82, 1.83, 1.73, 1.99, 1.85, 1.68, 1.87]
xbar = np.mean(x) # sample mean
mu0 = 1.75 # hypothesized value
s = np.std(x, ddof=1) # sample standard deviation
n = len(x) # sample size
tobs = (xbar - mu0) / (s / np.sqrt(n))
print(tobs)
```

```python
#tobs = 2.39687663116 # assume the t-value
tvalues = np.linspace(-10, 10, 100)
plt.plot(tvalues, stats.t.pdf(tvalues, n-1), 'b-', label="T(n-1)")
upper_tval_tvalues = tvalues[tvalues > tobs]
plt.fill_between(upper_tval_tvalues, 0, stats.t.pdf(upper_tval_tvalues, n-1), alpha = .6, label="p-value",color = 'red')
plt.legend()
```

## 7. Write a program to plot the following function: 𝑓(𝑥; 𝜃) = (1 + 𝜃)𝑥^𝜃, 0 < 𝑥 < 1, 𝜃 > −1.

```python
#defining f(x)
def f(x,teta):
    return((1+teta)*pow(x,teta))
def plotting_f(teta):
    if teta <= -1:
        print("wrong teta value..")
        return
    x = np.linspace(0,1,100)
    plt.grid(True)
    fig = plt.plot(x,f(x,teta))
    plt.title(f"plotting for theta = {teta}")
    plt.xlabel("x")
    plt.ylabel("y")
    print(fig)
```

```python
plotting_f(0.5)
```

```python
plotting_f(-0.5)
```

## 10. Generate random 100 integers. Write a program to find the mean, median, and standard deviation for the generated data.


```python
def mean(data,n):
    sum = 0
    for i in range(n):
        sum = data[i]
    return(sum/n)
```

```python
def median(data,n):
    if n%2 == 0:
        k = int(n/2)
        j = int(n/2)+1
        print('median =',data[k],data[j])
    else:
        k = n/2
        print('median =',data[k])
    return
```

```python
def sd(data,n,mu):
    sum = 0
    for i in range(n):
        sum = data[i]**2 - mu**2
    return(math.sqrt(sum))
```

```python
seed(1)
n = 100
data = randn(n)
data.sort()
print(data)
mu = mean(data,n)
sd = sd(data,n,mu)
print(f"\n\n\nFor the data the mean = {mu},\n standard deviation = {sd}")
median(data,n)
```
