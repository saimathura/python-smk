# Viusalizaton of Distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Plotting Binomial Function.

$f(x) = nCx p^x q^n-x$

```python

def plot_binomial(n,p):
    X = [i for i in range(n+1)]
    y = []
    for x in X:
        y.append(math.comb(n,x)*p**x*(1-p)**(n-x))
        
    plt.plot(X,y,color = 'red')
    plt.bar(X,y,color = 'green')
    plt.scatter(15,0.14)
    plt.title(f"Binomial Distribution for n = {n} p = {p}")
    plt.xlabel('x')
    plt.ylabel('f(x)')
```

```python
plot_binomial(30,0.5)
```

### Plotting Normal Function.

```python
def f(x,u,sig):
    pi = np.pi
    a = 1/math.sqrt(2*pi*sig**2)
    b = math.exp((-1/2)*(((x-u)/sig)**2))
    return a*b
def norm(u,sig):
    X=np.linspace(u-(4*sig),u+(4*sig),100)
    y = []
    for x in X:
        y.append(f(x,u,sig))
        plt.grid(True)
      
    #plt.bar(X,y,color = 'green')
    plt.plot(X,y,color = 'green')
    plt.scatter(u,f(u,u,sig),color = 'red')
    plt.title(f"Normal Distribution for mu = {u} sig = {sig}")
    plt.xlabel('x')
    plt.ylabel('f(x)')
```

```python
norm(0,1)
```

### Plotting Poisson Function.

```python
def f(lmbda,x):
    return ((math.exp(-lmbda)*math.pow(lmbda,x))/math.factorial(x))

def poi(lmbda):
    X=np.arange(0,30)
    y = []
    for x in X:
        y.append(f(lmbda,x))
        plt.grid(True)
      
    plt.bar(X,y,color = 'green')
    plt.plot(X,y,color = 'red')
    plt.title(f"Poisson Distribution for $\lambda$ = {lmbda} ")
    plt.xlabel('x')
    plt.ylabel('f(x)')
```

```python
poi(5)
```

### Plotting Gamma Function.

```python
def gam(alph):
    return math.factorial(alph-1)
def f(alph,lmbda,x):
    return ((math.pow(lmbda,alph)*math.pow(x,(alph-1))*math.exp(-lmbda*x))/gam(alph))
def gamma(alph,lmbda):
    X=np.linspace(0,20,100)
    y = []
    for x in X:
        y.append(f(alph,lmbda,x))
        plt.grid(True)
      
    #plt.bar(X,y,color = 'green')
    plt.plot(X,y,color = 'green')
    plt.title(f"Gamma Distribution for alpha  = {alph} $\lambda$ = {lmbda} ")
    plt.xlabel('x')
    plt.ylabel('f(x)')
```

```python
gamma(1,0.5)
```

### Plotting Exponential Function.

```python
def f(lmbda,x):
    return (lmbda*math.exp(-lmbda*x))
def exponential(lmbda):
    X=np.linspace(0,20,100)
    y = []
    for x in X:
        y.append(f(lmbda,x))
        plt.grid(True)
      
    #plt.bar(X,y,color = 'green')
    plt.plot(X,y,color = 'green')
    plt.title(f"Exponential Distribution for $\lambda$ = {lmbda} ")
    plt.xlabel('x')
    plt.ylabel('f(x)')
```

```python
exponential(1)
```

###  plotting geometric Distribution

```python
def f(p,x):
    return(p*math.pow((1-p),1-x))
def geometric(p):
    X=np.arange(0,10)
    y = []
    for x in X:
        y.append(f(p,x))
        plt.grid(True)
      
    plt.bar(X,y,color = 'green')
    plt.plot(X,y,color = 'red')
    plt.title(f"geometric Distribution for p = {p} x = {x}")
    plt.xlabel('x')
    plt.ylabel('f(x)')
```

```python
geometric(0.5)
```
